```python
def board():
    num=6
    for i in range(0,5):
        if(i%2==0):
            print('  |   | ')
            print(f'{brd[num]} | {brd[num+1]} | {brd[num+2]}')
            print('  |   | ')
            num-=3
        else:
            print('---------')
            
def play(brd,ch,player):
    while(True):
        print(f'Choose your position {player} 1-9')
        choice=int(input())-1
        if(choice<0 or choice >8):
            print('Invalid Choice')
        elif(brd[choice]!=' '):
            print(' Pos ALready filled')
        else:
            return choice
    
def check(brd,ch):
    temp=0
    for letter in brd:
        if(letter.isalpha()):
            temp+=1
    if(temp>=5):
        if(brd[0]==brd[1]==brd[2] and brd[0]!=' '):
            return(ch.index(brd[0])+1)
        elif(brd[0]==brd[4]==brd[8] and brd[0]!=' '):
            return(ch.index(brd[0])+1)
        elif(brd[0]==brd[3]==brd[6] and brd[0]!=' '):
            return(ch.index(brd[0])+1)
        elif(brd[1]==brd[4]==brd[7] and brd[1]!=' '):
            return(ch.index(brd[1])+1)
        elif(brd[2]==brd[4]==brd[6] and brd[2]!=' '):
            return(ch.index(brd[2])+1)
        elif(brd[2]==brd[5]==brd[8] and brd[2]!=' '):
            return(ch.index(brd[2])+1)
        elif(brd[3]==brd[4]==brd[5] and brd[3]!=' '):
            return(ch.index(brd[3])+1)
        elif(brd[6]==brd[7]==brd[8] and brd[6]!=' '):
            return(ch.index(brd[6])+1)
        elif(temp==9):
            return 3
        else:
            return 0 
    else:
        return 0
    
choice='Y'
while(choice=='Y'):
    ch=[]
    brd=[' ']*9
    player=[]
    choice='Y'
    print('Welcome to Tic Tac Toe')
    player.append(input('Enter Player 1 Name'))
    player.append(input('Enter Player 2 Name'))
    print('Enter X for playing first or Enter O for playing second')
    ch.append(input(f'{player[0]} Enter your choice? '))
    while(ch[0] not in ['X','O']):
        print('Wrong Choice')
        print('Enter X for playing first or Enter O for playing second')
        ch[0]=input(f'{player[0]} Enter your choice? ')
    pos=['X','O'].index(ch[0])
    if(ch[0]=='X'):
        ch.append('O')
    else:
        ch.append('X')
    win=0
    while(win==0 and choice=='Y'):
        board()
        brd[play(brd,ch,player[pos])]=ch[pos]
        win=check(brd,ch)
        if(win==0):
            if(pos==0):
                pos=1
            else:
                pos=0
            continue
        elif(win in [1,2]):
            board()
            print(f'Congratulations {player[pos]} ! You have won the game')
        elif(win==3):
            board()
            print('TIE')
        while(True):
            choice=input('Do you want to continue playing?Y/N: ')
            if(choice in ['Y','N']):
                break
print('Thank You for Playing Tic Tac Toe')
```
